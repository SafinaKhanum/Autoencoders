## **What are Autoencoders?**

An autoencoder is an unsupervised feed forward artificial neural network that learns to compress and encode the input data in order to reduce data dimensions and then learns how to reconstruct the data back from the reduced encoded form to a form that is as close to the original input as possible. 

Basically,
* It compresses or encodes the input data so the data dimensions are reduced
* Then it decompresses the encoded data and returns an output that is as close as possible to the original input provided

An autoencoder reduces dimensions by learning how to ignore the noise in the input data.

## **The 4 parts of an autoencoder**

1.   **Encoder** : Here the model learns to compress/encode the input data

2.   **Bottleneck** : This is the layer that contains the encoded input data in its lowest possible dimensions.

3. **Decoder**: Here the model learns to decode the compressed data such that the output data is as close as possible to the input data.

4. **Reconstruction loss** : This is the measure of how well the decoder is performing or how close the output data is to the original input data.


## **The flow of how an autoencoder works**

The below image clearly shows through an example, the flow of how an autoencoder works.

a1.png

**Step 1**: The auto encoder is fed an image of a digit. This is the input data that also consists of noise.

**Step 2** : The autoencoder comes up with a compressed/encoded representation of the image by getting rid of the noise and retaining only those important features of the image that are required.

**Step 3**: The autoencoder then decodes this compressed representation to return an output that is free of noise yet same as the input image.

**Step 4** : A reconstruction loss is then calculated to see how close the output image is to the original input image. Then a back propogation is done through the network to minimize this reconstruction loss as much as possible.



## **Key properties of autoencoders**

**1) Unsupervised / self supervised** :  Autoencoders are considered as an unsupervised learning technique because they dont need explicit labels to train on. They are infact self-supervised because they generate their own labels from the training data.

**2) Data specific** : Autoencoders work well on data that is similar to what they have been trained on. For example if an auto encoder is trained on images of digits then it can further work well in compressing only images of digits, if any other kind of image is passed through it, say the image of a t-shirt while the training was done on images of digits the autoencoder will not be able to compress it well and will result in a high reconstruction loss.

3) **No loss less compression** : The output of the autoencoder is not exactly the same as the input, it will be a close but degraded representation of it. 

## **The Architecture of autoencoders**

image.png

In an autoencoder both the encoder and decoder are fully-connected feedforward neural networks. From the above image, code is a single layer of an Artificial neural network with a dimensionality that is specified by the user. 

The decoder is usually the mirror image of the encoder in terms of layer structure.However there is no fixed rule that this condition has to be specified.The only requirement is that the input and output has to be the same in dimensionality. Everything else in the network are hyperparameters that can be specified by the user.

There are 4 hyperparameters that are used to train an autoencoder

**1) Code size** : It is the number of nodes in the middle layer. A smaller code size results in more compression.

**2) Number of layers** : There can be as many layers as a user wants in both the enocder and decoder.

**3) Number of nodes per layer** : The above image represents an architecture called the stacked autoencoder because the layers are stacked one after another. This looks like a sandwitch. With each subsequent layer of an encoder the number of nodes per layer decrease, and then increase back in the decoder. The decoder is symmetric to the encoder in the number of nodes per layer. 

**4) Loss function**: When the input data is in the range [0, 1] then a crossentropy loss funcction is used, else for numeric input data, the mean squared error loss function is used. 

A user can completely control the architecture of an autoencoder. An autoencoder can be made extremely powerfull increasing the number of layers, nodes per layer and the code size. This way the autoencoder can learn more complex codings. However the downside of making a very powerful autoencoder is overfitting. When too many layers or nodes are used to build an autoencoder, it will memorize the training data too well and perform well on it too. But it will perform very poorly on the test/unseen data.

Overfitting can also occur when the code size (or dimension of the bottleneck) is same as the dimension of the input dimension of th is the same as the input, or in the overcomplete case, where the dimension of the bottleneck is greater than the dimension of the input.

The stacked autoencoder architecture helps overcome this problem. Since the number of nodes in the code layer are less than the number of nodes in the input layer in this architecture, the autoencoder will not be able to exactly copy the inputs to the output, and will be forced to learn intelligent features. It will be forced to learn patterns from the input instead of memorizing the complete input.

Basically we are creating constraints on the code layer in this approach.In such a case an autoencoder is called undercomplete.

By training an undercomplete autoencoder, we force it to learn the most important features of the input/training data. 

Using a very small code size on the other hand will limit the capacity of the autoencoder to recreate the output as close as possible to the input.



## **The purpose of an autoencoder**
An autoencoder works similar to a feed forward artificial neural network except that in an autoencoder the purpose is to reconstruct its inputs instead of predicting the value of a dependent variable(Y).And since no labels are involved in case of an autoencoder it is categorized under unsupervised/self supervised learning.

The purpose of an autoencoder is not simply to copy the input to the output.

The intention behind using an autoencoder is to train the network such that the bottleneck/latent representation layer to learn some useful properties form the input data.

## **Types of autoencoders**

There are 7 different types of autoencoders :

**1) Denoising autoencoders**




image.png

When there are more nodes in the hidden layer than in the input, there is a chance that the autoencoder simply memorizes the input and generates the same output. This problem is overcome by using denoising autoencoders. Denoising autoencoders purposely introduce some noise and corrupt the input data. This is done by randomnly converting some of the input values to zero.These autoencoders take a partially corrupted input while training to recover the original undistorted input. The model learns a vector field for mapping the input data towards a lower dimensional manifold which describes the natural data to cancel out the added noise.


**2) Sparse autoencoders**

image.png

These type of autoencoders are ususally used for tasks like classification.Here the training process involves a "sparsity penalty".The sparsity constraint is applied on the hidden layer.Generally the loss function is constructed  by penalizing activations of hidden layers such that only a few nodes are activated when a single sample is fed into the network. This is done to ensure that the autoencoder is is actually learning latent representations instead of redundant information from the input data.Sparsity in the loss function, may be obtained during the training process, either by comparing the probability distribution of the hidden unit activations with some low desired value,or by manually zeroing all but the strongest hidden unit activations.


**3) Deep autoencoders**

image.png

As the name itself suggests, these type of autoencoders are called 'deep' due to more number of layers that are used to construct them. They consist of two symmetrical deep networks. One network for encoding and the other one for decoding. Usually there are 4-5 layers in the encoding network and another set of 4-5 networks in the decoding network.They are typically used for tasks liketopic modelling or dealing with data that is in the form of documents.


**4) Contractive autoencoder**

image.png

The purpose of a contractive autoencoder is to make the encoding process less sensitive to small variations in the training data set.This is done by adding a regularizer, or penalty term, to the cost function that the algorithm is trying to minimize. This autoencoder is similar to the sparse or denoising autoencoders except that it corresponds to the Frobenius norm of the Jacobian matrix of the encoder activations with respect to the input. Frobenius norm of the Jacobian matrix for the hidden layer is calculated with respect to input and it is basically the sum of square of all elements.


**5) Undercomplete autoencoder**

image.png

In this type of autoencoder the number of nodes in the code layer are less than the number of nodes in the input layer. This way the autoencoder is restricted from exactly copying the inputs to the output, and it is  forced to learn intelligent features. It is forced to learn patterns from the input instead of memorizing the complete input.Basically we are creating constraints on the code layer in this approach. The Objective is to minimize the loss function by penalizing the g(f(x)) for being different from the input x.Undercomplete autoencoders do not need any regularization as they maximize the probability of data rather than copying the input to the output.

**6) Convolutional autoencoder**

image.png

Autoencoders typically do not take into account the fact that a signal can be seen as a sum of other signals.Convolutional neural networks on the other hand learn to encode the input in a set of simple signals and then try to reconstruct the input from them.Here both the encoder and decoder networks are convolutional neural networks. They use the convolutional filters.Once these filters have been trained and they have learnt the patterns from the input data, they can be applied to any input in order to extract features. These features can then be used to do any task that requires a compressed/encoded representation of the input such as classification.


**7) Variational autoencoder**

image.png

In a variational autoencoder the encoder network is trained to return a probability distribution for describing each latent attribute or feature and not just one single value. Basically the learning/training happens by assigning a range of possible values to each feature(smile,skin colour etc) from the input data instead of assigning one single value like other autoencoders.Once the encoder comes up with a probability distribution for each feature, random samples from each latent probability distribution are picked to generate a vector as input for the decoder model.By doing so,we are enforcing a continuous, smooth latent space representation. For any sampling of the latent distributions, we expect the decoder model to be able to accurately reconstruct the input. Thus, values which are nearby to one another in latent space should correspond with very similar reconstructions.

## **Practical examples of an autoencoder**

## **Example 1 : Using an autoencoder for Anamoly detection**

When we have correlated input data at hand, the autoencoder method works very well for anamoly/outlier detection because the encoding operation relies on the correlated features to compress the data.

Data set - MNIST

```
#Importing all the required libraries and the mnist data set

import numpy as np
import keras
from keras.datasets import mnist
from keras.models import Sequential, Model
from keras.layers import Dense, Input, Conv2D, MaxPooling2D, UpSampling2D
from keras import optimizers
from keras.optimizers import Adam
import matplotlib.pyplot as plt
```

```
# Reshaping the data set and splitting it into train set and test set

(x_train, y_train), (x_test, y_test) = mnist.load_data()
train_x = x_train.reshape(60000, 784) / 255
val_x = x_test.reshape(10000, 784) / 255
```

```
#Shape of the train and test data set

x_train.shape,x_test.shape

#There are 60000 images in the train data and 10000 images in the test data. All are of 28 x 28 pixels
```

```
#Building a 6 layer sequential network

autoencoder = Sequential()
autoencoder.add(Dense(512,  activation='elu', input_shape=(784,)))
autoencoder.add(Dense(128,  activation='elu'))
autoencoder.add(Dense(10,    activation='linear', name="bottleneck"))
autoencoder.add(Dense(128,  activation='elu'))
autoencoder.add(Dense(512,  activation='elu'))
autoencoder.add(Dense(784,  activation='sigmoid'))
```

```
#Using mean squared error as the loss function and the adam optimizer to build the encoder

autoencoder.compile(loss='mean_squared_error', optimizer = Adam())
trained_model = autoencoder.fit(train_x, train_x, batch_size=100, epochs=10, verbose=1, validation_data=(val_x, val_x))
encoder = Model(autoencoder.input, autoencoder.get_layer('bottleneck').output)
encoded_data = encoder.predict(train_x)  # bottleneck/code layer representation
decoded_output = autoencoder.predict(train_x)        # reconstructing the compressed data
encoding_dim = 10

# Return the decoder
encoded_input = Input(shape=(encoding_dim,))
decoder = autoencoder.layers[-3](encoded_input)
decoder = autoencoder.layers[-2](decoder)
decoder = autoencoder.layers[-1](decoder)
decoder = Model(encoded_input, decoder)
```

The reconstruction loss for the validation set at the last epoch is 0.0112 . So as long as an image from the MNIST data set (the data set that the autencoder has been trained on) is passed the the reconstruction loss would we somewhere around this value. But when a completely random image that is not a part of the MNIST data set is passed the reconstruction loss will be very high indicating that the random image is an anamoly or outlier or very different from the kind of images that the model was trained on.

```
#Passing a random image to the model

%matplotlib inline
from keras.preprocessing import image
img = image.load_img("random_img.png", target_size=(28, 28), color_mode = "grayscale")
input_img = image.img_to_array(img)
inputs = input_img.reshape(1,784)
target_data = autoencoder.predict(inputs)
dist = np.linalg.norm(inputs - target_data, axis=-1)
print(dist)
```

The reconstruction loss for the above random image is 5174.32 which is very high when compared to the reconstruction loss on the validation set indicating that this image is an anamoly or an outlier.

## **Example 2 - Using an autoencoder for Denoising the data**

Data set : MNIST



```
# Loading the MNIST data set and splitting it into a train set and test set

(X_train, _), (X_test, _) = mnist.load_data()
X_train = X_train.astype('float32') / 255. #normalizing the data set
X_test = X_test.astype('float32') / 255.
```

```
#Reshaping the train and test set

X_train = X_train.reshape(-1, 28,28, 1)
X_test = X_test.reshape(-1, 28,28, 1)
X_train.shape, X_test.shape
```

```
# Building the Convolutional autoencoder - auntoencoder with a convolutional neural network 
autoencoder = Sequential()
autoencoder.add(Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(28, 28, 1)))
autoencoder.add(MaxPooling2D((2, 2), padding='same'))
autoencoder.add(Conv2D(8, (3, 3), activation='relu', padding='same'))
autoencoder.add(MaxPooling2D((2, 2), padding='same'))
autoencoder.add(Conv2D(8, (3, 3), activation='relu', padding='same'))
 
# Encoding the input data
autoencoder.add(MaxPooling2D((2, 2), padding='same'))
 
autoencoder.add(Conv2D(8, (3, 3), activation='relu', padding='same'))
autoencoder.add(UpSampling2D((2, 2)))
autoencoder.add(Conv2D(8, (3, 3), activation='relu', padding='same'))
autoencoder.add(UpSampling2D((2, 2)))
autoencoder.add(Conv2D(32, (3, 3), activation='relu'))
autoencoder.add(UpSampling2D((2, 2)))
autoencoder.add(Conv2D(1, (3, 3), activation='relu', padding='same'))
```

```
# Using mean squared error as the loss function and the adam optimizer 
autoencoder.compile(optimizer='adam', loss='mean_squared_error')
```

```
# Introducing random noise into the images

X_train_noisy = X_train + 0.25 * np.random.normal(size=X_train.shape) #adding noise (25%of random values - noise to the actual image)
X_test_noisy = X_test + 0.25 * np.random.normal(size=X_test.shape) #adding noise
 
#normalizing the data so that all values are between 0 and 1

X_train_noisy = np.clip(X_train_noisy, 0., 1.)
X_test_noisy = np.clip(X_test_noisy, 0., 1.)
```

```
# Fitting the model

autoencoder.fit(X_train_noisy, X_train, epochs=3, batch_size=100, shuffle=True, validation_split=0.2) 
```

```
#Plotting the noisy images and the corresponding denoised images

decoded_imgs = autoencoder.predict(X_test_noisy)

plt.figure(figsize=(20, 4))
for i in range(10):
    # original
    plt.subplot(2, 10, i + 1)
    plt.imshow(X_test_noisy[i].reshape(28, 28))
    plt.gray()
    plt.axis('off')

    # reconstruction
    plt.subplot(2, 10, i + 1 + 10)
    plt.imshow(decoded_imgs[i].reshape(28, 28))
    plt.gray()
    plt.axis('off')

plt.tight_layout()
plt.show()
```

The first row in the above output shows images with the randomnly introduced noise (input data) and the second row shows the denoised images (output data) which are comapritively more clear.
